```python
from kpl_layers import polynomial
import torch
```

```python
pnn = polynomial.PolynomialLayer(1,2,1)
print(pnn)
```

```python
print(pnn.state_dict())
```

```python
# Create some random data to train on
n_train = 100000
x_train = 20*torch.rand(n_train,1) - 10
y_train = -0.1*x_train*x_train + 0.7*x_train - 2
ds_train = torch.utils.data.TensorDataset(x_train,y_train)
loader_train = torch.utils.data.DataLoader(ds_train,batch_size=100)

n_test = 1000
x_test = 20*torch.rand(n_test,1) - 10
y_test = -0.1*x_test*x_test + 0.7*x_test - 2
ds_test = torch.utils.data.TensorDataset(x_test,y_test)
loader_test = torch.utils.data.DataLoader(ds_test,batch_size=100)
```

```python
# Configure the network for training
loss = torch.nn.MSELoss()
optimizer = torch.optim.Adam(pnn.parameters(),lr=1e-4)
```

```python
# Taken from the pytorch tutorial code
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    for batch, (X, y) in enumerate(dataloader):
        # Compute prediction and loss
        pred = model(X)
        loss = loss_fn(pred, y)

        # Backpropagation
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 200 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")


def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Avg loss: {test_loss:>8f} \n")
```

```python
epochs = 30
for t in range(epochs):
    print(f"Epoch {t+1}\n-------------------------------")
    train_loop(loader_train, pnn, loss, optimizer)
    test_loop(loader_test, pnn, loss)
print("Done!")
```
